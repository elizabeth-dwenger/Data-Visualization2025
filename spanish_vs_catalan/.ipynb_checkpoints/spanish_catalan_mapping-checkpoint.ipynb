# Creating Map for Spanish vs Catalan

```python
import osmnx as ox
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Download Catalonia boundary
catalonia = ox.geocode_to_gdf("Catalonia, Spain")
catalonia
```

```python
tags = {"boundary": "administrative", "admin_level": ["6", "7"]}

regions = ox.features_from_polygon(catalonia.geometry.iloc[0], tags)

# Keep only polygons
regions = regions[regions.geometry.type.isin(["Polygon", "MultiPolygon"])]
regions.head()
```

```python
print(regions.columns)
print(regions["name"].unique())
```

```python
df = pd.read_csv("spanish_catalan.csv")

# Keep only Catalan / Spanish / Catalan+Spanish
df_sub = df[df["habitual language"].isin(["Catalan","Spanish","Catalan and Spanish"])]

# Pivot table
df_pivot = df_sub.pivot(index="territorial area of ​​stratification", columns="habitual language", values="value").reset_index()
df_pivot.head()
```

```python
tags = {"boundary": "administrative", "admin_level": "7"} # Level 7 = Comarcas
regions_gdf = ox.features.features_from_place("Catalonia, Spain", tags)
regions_gdf = regions_gdf[regions_gdf.geometry.type.isin(["Polygon", "MultiPolygon"])]
```

```python
gdf_merged = regions_gdf.merge(df_pivot, left_on="name", right_on="region", how="left")
```

```python
fig, ax = plt.subplots(figsize=(8,10))
gdf_merged.plot(
    column="Catalan",
    cmap="Blues",
    edgecolor="black",
    linewidth=0.8,
    legend=True,
    ax=ax
)
ax.set_title("Habitual Use of Catalan by Region (thousands)", fontsize=14)
ax.axis("off")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,10))
gdf_merged.plot(
    column="Spanish",
    cmap="Reds",
    edgecolor="black",
    linewidth=0.8,
    legend=True,
    ax=ax
)
ax.set_title("Habitual Use of Spanish by Region (thousands)", fontsize=14)
ax.axis("off")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,10))
gdf_merged.plot(
    column="Catalan and Spanish",
    cmap="Greens",
    edgecolor="black",
    linewidth=0.8,
    legend=True,
    ax=ax
)
ax.set_title("Habitual Use of Catalan & Spanish by Region (thousands)", fontsize=14)
ax.axis("off")
plt.show()
```

```python
df = place_of_birth.copy()

df = df[df["concept"].str.contains("proportion", case=False, na=False)]

# Focus on full proficiency
df = df[df["language skills"] == "knowledge of all skills"]

# Keep only Catalan & Spanish
df = df[df["language"].isin(["Catalan", "Spanish"])]
```

```python
df
```

```python
pivot = df.pivot_table(
    index="place of birth",
    columns="language",
    values="value"
)

pivot = pivot.sort_index()

pivot
```

```python
ax = pivot.plot(kind="bar", figsize=(8,5))

ax.set_ylabel("Proportion with full language proficiency (%)")
ax.set_xlabel("Place of birth")
ax.set_title("Full Language Proficiency by Place of Birth (Barcelona, 2023)")

ax.legend(title="Language")
ax.set_ylim(0, 100)

plt.tight_layout()
plt.show()
```

```python
def prep_percent_data(df, outside_label):
    out = df.copy()
    
    # Keep ONLY percentage rows
    out = out[
        out["concept"].str.contains("proportion", case=False, na=False)
    ]
    
    # Keep full proficiency
    out = out[out["language skills"] == "can understand"]
    
    # Keep Catalan only
    out = out[out["language"] == "Catalan"]
    
    # Clean labels
    out["place of birth"] = out["place of birth"].replace(outside_label)
    
    # Ensure numeric
    out["value"] = pd.to_numeric(out["value"], errors="coerce")
    
    return out[["place of birth", "value"]]
```

```python
place_birth_pct = prep_percent_data(
    place_of_birth,
    {"rest of Spain": "Born outside Catalonia"}
)

father_birth_pct = prep_percent_data(
    father,
    {"Spain": "Father born outside Catalonia"}
)
```

```python
# 1. Define the Raw Data (English labels)
# N_total = 6000, N_catalan = 2010, N_spanish = 3339
data = {
    'district': [
        'CIUTAT VELLA', 'EIXAMPLE', 'SANTS - MONTJUÏC', 'LES CORTS', 
        'SARRIÀ - SANT GERVASI', 'GRÀCIA', 'HORTA - GUINARDÓ', 
        'NOU BARRIS', 'SANT ANDREU', 'SANT MARTÍ'
    ],
    'pop_pct': [6.1, 16.7, 11.3, 4.9, 8.7, 7.6, 10.7, 10.4, 9.1, 14.6],
    'cat_pct_of_total': [3.6, 18.5, 11.1, 6.6, 12.3, 9.9, 10.9, 5.1, 8.8, 13.3],
    'spa_pct_of_total': [6.2, 15.0, 10.8, 4.3, 6.6, 5.9, 11.3, 14.3, 10.2, 15.3]
}

df = pd.DataFrame(data)

# Convert to "Internal District %"
# Logic: (Dist_Cat_Count / Dist_Total_Pop) * 100
total_n = 6000
cat_n = 2010
spa_n = 3339

df['Catalan Percentage'] = (df['cat_pct_of_total'] * cat_n) / (df['pop_pct'] * total_n) * 100
df['Spanish Percentage'] = (df['spa_pct_of_total'] * spa_n) / (df['pop_pct'] * total_n) * 100

# Fetch Barcelona Boundaries from OSM
districts_gdf = ox.features_from_place("Barcelona, Spain", tags={"admin_level": "9"})
districts_gdf['name_upper'] = districts_gdf['name'].str.upper().replace({
    'SANTS-MONTJUÏC': 'SANTS - MONTJUÏC',
    'SARRIÀ-SANT GERVASI': 'SARRIÀ - SANT GERVASI',
    'HORTA-GUINARDÓ': 'HORTA - GUINARDÓ'
})

merged = districts_gdf.merge(df, left_on='name_upper', right_on='district')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(22, 12), facecolor='#ffffff')

# Settings for a cleaner look
maps_to_plot = [('Catalan Percentage', 'YlOrRd', axes[0]), ('Spanish Percentage', 'YlGnBu', axes[1])]

for col, cmap, ax in maps_to_plot:
    # Plot the choropleth
    merged.plot(column=col, cmap=cmap, legend=True, ax=ax, 
                edgecolor='0.3', linewidth=0.8, legend_kwds={'shrink': 0.6})
    
    # Add percentage labels to the center of each district
    for idx, row in merged.iterrows():
        # Get centroid for text placement
        centroid = row.geometry.centroid
        ax.text(centroid.x, centroid.y, f"{row[col]:.1f}", 
                fontsize=9, fontweight='bold', ha='center', color='black',
                bbox=dict(facecolor='white', alpha=0.5, edgecolor='none', boxstyle='round,pad=0.2'))

    ax.set_title(f'Habitual Language - {col}', fontsize=18, pad=20, fontweight='bold')
    ax.axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.suptitle('Language Dominance within Barcelona Districts', fontsize=26, fontweight='black')

# Save and Show
plt.savefig('barcelona_language_map.png', dpi=300, bbox_inches='tight')
plt.show()
```
