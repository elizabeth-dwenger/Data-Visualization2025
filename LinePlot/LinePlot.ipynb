# Tech Stock Performance: NVIDIA vs. Intel

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.interpolate
from matplotlib import font_manager
import matplotlib.dates as mdates
from matplotlib.dates import date2num
from matplotlib.font_manager import FontProperties
```

```python
personal_path = '../../Fonts_GT_Super/Desktop'
font_path = personal_path + '/GT-Eesti/GT-Eesti-Text-Book-Trial.otf'
font = FontProperties(fname=font_path)
bold_font = FontProperties(fname=personal_path + '/GT-Eesti/GT-Eesti-Display-Bold-Trial.otf')
```

```python
nvidia = pd.read_csv("nvda.csv", parse_dates=['Date'])
intel = pd.read_csv("intc.csv", parse_dates=['Date'])
```

```python
nvidia.head()
```

```python
start_date = pd.to_datetime('2021-01-01')
nvidia = nvidia[nvidia['Date'] >= start_date].sort_values('Date')
intel = intel[intel['Date'] >= start_date].sort_values('Date')
```

```python
for df in [nvidia, intel]:
    df['Close'] = df['Close/Last'].replace('[\$,]', '', regex=True).astype(float)
    df['Date_num'] = mdates.date2num(df['Date'])  # Convert dates to numeric format
```

```python
# Create smoothing splines
def prepare_spline(df):
    x = np.linspace(df['Date_num'].min(), df['Date_num'].max(), 300)
    spline = scipy.interpolate.make_interp_spline(df['Date_num'], df['Close'], k=3)
    return x, spline(x)
```

```python
x_smooth, nvidia_smooth = prepare_spline(nvidia)
_, intel_smooth = prepare_spline(intel)
```

```python
bold_title = bold_font.copy()
bold_title.set_size(18) 
```

```python
plt.figure(figsize=(12, 7), dpi=500)
ax = plt.gca()

# Fill between areas
ax.fill_between(x_smooth, nvidia_smooth, intel_smooth, 
                where=nvidia_smooth > intel_smooth, 
                color='#74B101', alpha=0.3, interpolate=True)
ax.fill_between(x_smooth, nvidia_smooth, intel_smooth, 
                where=intel_smooth > nvidia_smooth, 
                color='#0A5DA3', alpha=0.3, interpolate=True)

ax.plot(x_smooth, nvidia_smooth, color='#74B101', linewidth=2.5, label='NVIDIA')
ax.plot(x_smooth, intel_smooth, color='#0A5DA3', linewidth=2.5, label='Intel')

# Format dates on x-axis
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.xaxis.set_major_locator(mdates.YearLocator())
plt.xticks()

title_text = [
    ("AI Chip Race ", 'black'),
    ("NVIDIA", '#74B101'),
    (" vs ", 'black'),
    ("Intel", '#0A5DA3')
]

current_x = 0.3 
i = 0.015
for text, color in title_text:
    i -=0.003
    t = ax.text(current_x, 1.05, text, 
                transform=ax.transAxes, 
                fontproperties=bold_title,
                color=color,
                ha='left')
    
    # Get rendered text width and update position
    plt.draw()
    bbox = t.get_window_extent()
    current_x += bbox.width / ax.figure.dpi / ax.figure.get_size_inches()[0]+i  # Convert pixels to Axes coordinates

events = [
    # (Text, Event Date, Target Company, Text Position (x_date, y_value))
    ('2021-02\nIntel CEO Change', '2021-02-15', 'intel', ('2021-04-01', 80)),
    ('2022-09\nRTX 4000 Launch', '2022-09-20', 'nvidia', ('2022-07-01', 60))
]

for text, event_date, target, text_pos in events:
    # Convert all dates to numeric values
    event_num = date2num(pd.to_datetime(event_date))
    text_x_num = date2num(pd.to_datetime(text_pos[0]))
    
    # Find y-value on the actual stock curve
    if target == 'nvidia':
        y_val = nvidia_smooth[np.abs(x_smooth - event_num).argmin()]
    else:
        y_val = intel_smooth[np.abs(x_smooth - event_num).argmin()]
    
    ax.annotate(text, 
                xy=(event_num, y_val),
                xytext=(text_x_num, text_pos[1]),  # Use date-based x-coordinate
                arrowprops=dict(
                    arrowstyle="->", 
                    color='#74B101' if target == 'nvidia' else '#0A5DA3',
                    connectionstyle="arc3,rad=-0.2"
                ),
                fontsize=9,
                color='#74B101' if target == 'nvidia' else '#0A5DA3',
                ha='center',
                va='center',
                fontproperties=font)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.yaxis.grid(True, alpha=0.2)
ax.set_ylabel("Stock Price", font=font)
ax.set_xlabel("Year", font=font)
plt.legend(frameon=False, loc='upper left', prop=font)

for tick in ax.get_xticklabels():
    tick.set_fontproperties(font)
for tick in ax.get_yticklabels():
    tick.set_fontproperties(font) 

plt.tight_layout()
plt.savefig('chip_stocks_comparison.png', bbox_inches='tight', dpi=500)
plt.show()
```

```python
np.random.seed(0)
x = np.linspace(0, 10, 100)
y1 = np.sin(x) + np.random.normal(0, 0.2, len(x)) + 1
y2 = np.cos(x) + np.random.normal(0, 0.2, len(x)) + 1
```

```python
smooth_x = np.linspace(x.min(), x.max(), 300)
smooth_y1 = scipy.interpolate.make_interp_spline(x, y1, k=3)(smooth_x)
smooth_y2 = scipy.interpolate.make_interp_spline(x, y2, k=3)(smooth_x)

fig, ax = plt.subplots(figsize=(10, 6), dpi=500)

ax.fill_between(smooth_x, smooth_y1, smooth_y2, where=smooth_y1 > smooth_y2, 
                color='#F37C6A', alpha=0.3, interpolate=True)
ax.fill_between(smooth_x, smooth_y1, smooth_y2, where=smooth_y2 > smooth_y1, 
                color='#66BEFD', alpha=0.3, interpolate=True)

ax.plot(smooth_x, smooth_y1, color='#F37C6A', linewidth=2, label="Red States")
ax.plot(smooth_x, smooth_y2, color='#66BEFD', linewidth=2, label="Blue States")

title_x, title_y = 0.3, 1.05 
ax.text(title_x, title_y, "Words", font=bold_title, ha='center', transform=ax.transAxes)
ax.text(title_x + 0.1, title_y, "red", font=bold_title, color='#F37C6A', ha='center', transform=ax.transAxes)
ax.text(title_x + 0.2, title_y, "vs.", font=bold_title, ha='center', transform=ax.transAxes)
ax.text(title_x + 0.3, title_y, "blue", font=bold_title, color='#66BEFD', ha='center', transform=ax.transAxes)

# Subtitle
subtitle_x, subtitle_y = 0.5, 1.01  
ax.text(subtitle_x, subtitle_y, "Words",
        fontsize=12, ha='center', transform=ax.transAxes, color='gray', font=font)

ax.annotate("Red", xy=(7, 1.8), xytext=(5, 2.2),
             arrowprops=dict(arrowstyle="->", color='#F37C6A'), fontsize=12, color='#F37C6A', font=font)

ax.annotate("Blue", xy=(2, 1.8), xytext=(4, 2.2),
             arrowprops=dict(arrowstyle="->", color='#66BEFD'), fontsize=12, color='#66BEFD', font=font)

ax.spines['left'].set_visible(False) 
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.yaxis.grid(True, linestyle='-', alpha=0.4, color='gray') 
ax.xaxis.grid(False)  

ax.set_xlabel("Time", fontsize=12, font=font)
ax.set_ylabel("y-axis", fontsize=12, font=font)

for tick in ax.get_xticklabels():
    tick.set_fontproperties(font)
for tick in ax.get_yticklabels():
    tick.set_fontproperties(font) 

plt.show()
```
