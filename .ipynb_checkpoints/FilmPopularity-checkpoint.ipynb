# Popularity of Estonian-Language Films I watched as a child

- Leiutajateküla Lotte (Lotte from Gadgetville)
- Mõmmi ja aabits
- Muumioru Lood/Muumitrollid (Moomin)
- Krokodill Gena ja sõbrad (Cheburashka/Чебурашка)

Recently I was thinking about some of the films I watched as a child; many of them were in Estonian. Mõmmi ja aabits and Krokodill Gena ja sõbrad were released in during the Soviet occupation of Estonia (so was Muumioru Lood, but this is Finnish), and my memories of these films are that they are somewhat depressing/strange. I was curious what Google Trends said about their popularity over the years, so I decided to visualize these. Some of these films are not originally Estonian, but I watched them as a child in Estonian. 

```python
import re
import matplotlib.pyplot as plt
from matplotlib import font_manager
from matplotlib.font_manager import FontProperties
import pandas as pd
```

### Data Pre-processing: Films in their Estonian Name

```python
eesti = pd.read_csv('eesti_keeles.csv', delimiter=',', skiprows=1)
eesti
```

```python
eesti.columns = [re.sub(r":?\s*\(Worldwide\)", "", col) for col in eesti.columns]
```

```python
eesti["Month"] = pd.to_datetime(eesti["Month"])
```

```python
eesti.set_index("Month", inplace=True)
```

```python
for col in eesti.columns[1:]: 
    eesti[col] = pd.to_numeric(eesti[col], errors="coerce")
```

```python
eesti.fillna(1, inplace=True)
eesti
```

### Data Pre-processing: Films in their English Name

```python
english = pd.read_csv('english_names.csv', delimiter=',', skiprows=1)
```

```python
english.columns = [re.sub(r":?\s*\(Worldwide\)", "", col) for col in english.columns]
```

```python
english["Month"] = pd.to_datetime(english["Month"])
```

```python
english.set_index("Month", inplace=True)
```

```python
for col in english.columns: 
    english[col] = pd.to_numeric(english[col], errors="coerce")
```

```python
english.fillna(1, inplace=True)
english
```

### Plotting

```python
personal_path = '../Fonts_GT_Super/Desktop'
font_path = personal_path + '/GT-Super/GT-Super-Display-Regular-Trial.otf'
font = FontProperties(fname=font_path)
bold_font = FontProperties(fname=personal_path + '/GT-Super/GT-Super-Display-Bold-Trial.otf')
sansserif = FontProperties(fname=personal_path + '/GT-Eesti/GT-Eesti-Display-Regular-Trial.otf')
```

```python
custom_colors = {
    "Leiutajateküla Lotte": "#003049", 
    "Mõmmi ja aabits": "#780000",    
    "Muumioru Lood": "#c1121f",    
    "Krokodill Gena": "#669bbc"  
}

colors_second = {
    "Lotte from Gadgetville": "#003049",
    "Mõmmi ja aabits": "#780000",
    "Moomin": "#c1121f",
    "Cheburashka": "#669bbc" 
}
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 6), dpi=300)

### Plot Estonian

for col in eesti.columns:
    color = custom_colors.get(col) 
    axes[0].plot(eesti.index, eesti[col], label=col, color=color) 

axes[0].set_xlabel("Month", fontproperties=font, fontsize=12)  
axes[0].set_ylabel("Value", fontproperties=font, fontsize=12) 
axes[0].set_title("Google Trend Data of My Childhood Films in their Estonian Name", fontproperties=sansserif, fontsize=18) 

axes[0].legend(prop=font)
axes[0].tick_params(axis="x", labelsize=10)
axes[0].tick_params(axis="y", labelsize=10)
axes[0].grid(False) 
axes[0].spines["top"].set_visible(False)
axes[0].spines["right"].set_visible(False)
axes[0].set_xlim(eesti.index.min(), eesti.index.max())
axes[0].set_ylim(eesti.min().min(), eesti.max().max())

for tick in axes[0].get_xticklabels():
    tick.set_fontproperties(font)
for tick in axes[0].get_yticklabels():
    tick.set_fontproperties(font) 


### Plot English

for col in english.columns:
    color = colors_second.get(col) 
    axes[1].plot(english.index, english[col], label=col, color=color)

axes[1].set_xlabel("Month", fontproperties=font, fontsize=12)  
axes[1].set_ylabel("Value", fontproperties=font, fontsize=12) 
axes[1].set_title("Google Trend Data of My Childhood Films in their English Name", fontproperties=sansserif, fontsize=18)

axes[1].legend(prop=font)
axes[1].tick_params(axis="x", labelsize=10)
axes[1].tick_params(axis="y", labelsize=10)
axes[1].grid(False)
axes[1].spines["top"].set_visible(False)
axes[1].spines["right"].set_visible(False)
axes[1].set_xlim(english.index.min(), english.index.max())
axes[1].set_ylim(english.min().min(), english.max().max())

for tick in axes[1].get_xticklabels():
    tick.set_fontproperties(font)
for tick in axes[1].get_yticklabels():
    tick.set_fontproperties(font) 

plt.tight_layout()
plt.show()
```
